## Analyse des Pays et Institutions

### Subtask:
Identifier les 10 pays et institutions ayant le plus grand nombre de scientifiques dans le dataset, et potentiellement analyser la moyenne du 'c score' par pays/institution pour évaluer leur impact global.


**Reasoning**:
The first step is to identify the top 10 countries based on the number of scientists and then calculate their average 'c score'. I will group the DataFrame by 'Country', count the number of unique scientists (assuming each row is a unique scientist for simplicity in this context), and then calculate the mean 'c score' for each country. Finally, I will sort these by the count of scientists and select the top 10.



```python
df_main = df['0,1']

# 1. Count the number of scientists per 'Country' and get the top 10 countries
country_counts = df_main['Country'].value_counts()
top_10_countries = country_counts.nlargest(10).index

# 2. For these 10 countries, calculate the mean 'c score'
country_c_score_mean = df_main[df_main['Country'].isin(top_10_countries)].groupby('Country')['c score'].mean()

# Combine the counts and mean c-scores for the top 10 countries
country_analysis = pd.DataFrame({
    'Nombre de Scientifiques': country_counts[top_10_countries],
    'Moyenne du c score': country_c_score_mean
}).sort_values(by='Nombre de Scientifiques', ascending=False)

# 3. Display the top 10 countries
print("Top 10 pays par nombre de scientifiques et leur c score moyen:\n")
print(country_analysis)
```

**Reasoning**:
Now, I will identify the top 10 institutions based on the number of scientists and calculate their average 'c score'. I will group the DataFrame by 'Institute Name', count the number of unique scientists, and then calculate the mean 'c score' for each institution. Finally, I will sort these by the count of scientists and select the top 10.



```python
df_main = df['0,1']

# 4. Count the number of scientists per 'Institute Name' and get the top 10 institutions
institute_counts = df_main['Institute Name'].value_counts()
top_10_institutes = institute_counts.nlargest(10).index

# 5. For these 10 institutions, calculate the mean 'c score'
institute_c_score_mean = df_main[df_main['Institute Name'].isin(top_10_institutes)].groupby('Institute Name')['c score'].mean()

# Combine the counts and mean c-scores for the top 10 institutions
institute_analysis = pd.DataFrame({
    'Nombre de Scientifiques': institute_counts[top_10_institutes],
    'Moyenne du c score': institute_c_score_mean
}).sort_values(by='Nombre de Scientifiques', ascending=False)

# 6. Display the top 10 institutions
print("Top 10 institutions par nombre de scientifiques et leur c score moyen:\n")
print(institute_analysis)
```

## Préparer les Données des Chercheurs par Pays

### Subtask:
Calculer le nombre de scientifiques par pays à partir du DataFrame `df['0,1']`. Utiliser la colonne 'Country' (abréviation) pour le groupement, et stocker ce résultat dans un nouveau DataFrame nommé `scientists_per_country`.


**Reasoning**:
To calculate the number of scientists per country, I will group the main DataFrame by the 'Country' column, count the entries in each group using `.size()`, convert the series to a DataFrame, and then rename the column to 'Nombre de Scientifiques'.



```python
df_main = df['0,1']

# Group by 'Country' and count the number of scientists
scientists_per_country = df_main.groupby('Country').size().reset_index(name='Nombre de Scientifiques')

print("Nombre de scientifiques par pays :\n")
print(scientists_per_country.head())
```

## Standardiser les Noms de Pays

### Subtask:
Convertir les abréviations de pays dans le DataFrame `scientists_per_country` en noms de pays complets en utilisant `df_key['Keys']`.


**Reasoning**:
I will merge the `scientists_per_country` DataFrame with `df_key['Keys']` to replace country abbreviations with full names, then select and rename the necessary columns to finalize the `scientists_per_country` DataFrame with standardized country names.



```python
df_keys_df = df_key['Keys']

# Merge scientists_per_country with df_keys_df to get full country names
scientists_per_country = pd.merge(
    scientists_per_country,
    df_keys_df[['Country', 'Abbreviation']],
    left_on='Country',
    right_on='Abbreviation',
    how='left'
)

# Drop the 'Abbreviation' column and the old 'Country' abbreviation column
scientists_per_country.drop(columns=['Country_x', 'Abbreviation'], inplace=True)

# Rename the new 'Country_y' column (full country name) to 'Country'
scientists_per_country.rename(columns={'Country_y': 'Country'}, inplace=True)

print("DataFrame 'scientists_per_country' avec noms de pays standardisés :")
print(scientists_per_country.head())
```

## Fusionner les Données

### Subtask:
Fusionner le DataFrame des chercheurs par pays (avec noms complets) avec le DataFrame `df_scimago` sur les noms de pays pour avoir toutes les informations nécessaires pour l'analyse dans un seul DataFrame.


**Reasoning**:
To merge the DataFrames, I will first rename the country column in `df_scimago` to 'Country' for consistency, then perform a left merge with `scientists_per_country` on the 'Country' column to create `merged_data_country`.



```python
df_scimago_cleaned = df_scimago.rename(columns={'Unnamed: 0': 'Country'})

# Perform a left merge to keep all countries from scientists_per_country
merged_data_country = pd.merge(
    scientists_per_country,
    df_scimago_cleaned,
    on='Country',
    how='left'
)

print("Premières lignes du DataFrame fusionné 'merged_data_country' :\n")
print(merged_data_country.head())
```

## Analyser et Visualiser les Corrélations

### Subtask:
Calculer les coefficients de corrélation de Pearson entre le 'Nombre de Scientifiques' et les métriques clés de `df_scimago` ('Documents', 'Citations', etc.). Visualiser ces relations à l'aide de nuages de points (scatter plots).


**Reasoning**:
I will define the list of relevant columns from `df_scimago` and then calculate the Pearson correlation coefficient between 'Nombre de Scientifiques' and each of these columns using the `merged_data_country` DataFrame. Finally, I will display the calculated correlations.



```python
scimago_metrics = [
    'Documents',
    'Citable documents',
    'Citations',
    'Self-citations',
    'Citations per document',
    'International Collaboration (%)',
    'International Collaboration Impact',
    'Academic-Corporate Collaboration (%)',
    'Academic-Corporate Collaboration Impact',
    'Output in Top 10% Citation Percentiles (%)',
    'Publications in Q1 Journal Quartile by SJR (%)',
    'Patent-Citations per Scholarly Output'
]

# Calculate Pearson correlation coefficients
correlations = {}
for metric in scimago_metrics:
    if metric in merged_data_country.columns:
        correlation = merged_data_country['Nombre de Scientifiques'].corr(merged_data_country[metric])
        correlations[metric] = correlation

# Display the correlations
print("Coefficients de corrélation de Pearson avec 'Nombre de Scientifiques':\n")
for metric, corr_value in correlations.items():
    print(f"- {metric}: {corr_value:.2f}")
```

**Reasoning**:
I will create a scatter plot for each `scimago_metric` against 'Nombre de Scientifiques' to visualize their relationships. Each plot will have a title and appropriate axis labels and will be displayed separately.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create scatter plots for each metric
for metric in scimago_metrics:
    if metric in merged_data_country.columns:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(data=merged_data_country, x='Nombre de Scientifiques', y=metric, alpha=0.6)
        plt.title(f'Relation entre Nombre de Scientifiques et {metric}')
        plt.xlabel('Nombre de Scientifiques')
        plt.ylabel(metric)
        plt.grid(True)
        plt.show()
```

```python
import pandas as pd
df_world_bank = pd.read_excel("world_bank_data.xlsx")
```

```python
import numpy as np

df_world_bank_cleaned = df_world_bank.rename(columns={'Country Name': 'Country'})

# Identify numerical columns in df_world_bank_cleaned
numerical_cols_world_bank = df_world_bank_cleaned.select_dtypes(include=np.number).columns.tolist()

# Exclude 'Number of scientists' from the list of columns to correlate against themselves
if 'Number of scientists' in numerical_cols_world_bank:
    numerical_cols_world_bank.remove('Number of scientists')

results = []

for col in numerical_cols_world_bank:
    # Create a temporary DataFrame, dropping rows with NaNs in either column
    temp_df = df_world_bank_cleaned[['Number of scientists', col]].dropna()

    # Get the number of valid data points (N)
    n_data_points = len(temp_df)

    # Only compute correlation if N >= 2 and both columns have non-zero standard deviation
    if n_data_points >= 2 and temp_df['Number of scientists'].std() > 0 and temp_df[col].std() > 0:
        correlation_r = temp_df['Number of scientists'].corr(temp_df[col])
        results.append({'Metric': col, 'R': correlation_r, 'N': n_data_points})

filtered_correlations_df = pd.DataFrame(results)

# Sort by absolute value of R in descending order
if not filtered_correlations_df.empty:
    filtered_correlations_df['Absolute R'] = filtered_correlations_df['R'].abs()
    filtered_correlations_df = filtered_correlations_df.sort_values(by='Absolute R', ascending=False).drop(columns=['Absolute R'])

print("Corrélations entre 'Number of scientists' et les indicateurs de la Banque Mondiale (R et N) :\n")
print(filtered_correlations_df.to_string())
```
